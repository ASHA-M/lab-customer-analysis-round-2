```python
def clean(file):
    file = file.drop(['Unnamed: 0'], axis=1) #1
    file['State'].value_counts(dropna=False) #2
    file['Vehicle Type'] = file['Vehicle Type'].fillna('not A') #4
    file['Vehicle Type'].value_counts(dropna=False) #5
    file = file[file['State'].isna()==False] #7
    file = file[file['Months Since Last Claim'].isna()==False] #8
    file = file[file['Vehicle Class'].isna()==False] #9
    file['Effective To Date'] = pd.to_datetime(file['Effective To Date']) #10
    file['Effective To Date Month'] = file['Effective To Date'].dt.month #11
    file[file['Effective To Date Month'].isin([1, 2, 3])] #12
    columns = [] #13
    for i in file.columns:
        columns.append(i.lower())
    file.columns = columns
    return file

```

##### Creating bins for total claim amount

```python
# we first describe the total claim amount so we can get the different percentile values
file['total claim amount'].describe()
```

```python
# we use the describe values to create a function which will help put the row values in total claim amount to different bins
def c_bins(row):
    if row < 272:
        return ' < 25%'
    elif row < 383:
        return '25%-50%'
    elif row < 548:
        return '50%-75%'
    else:
        return '> 75%'
# go and get every row of total claim amount and apply the bins() function to every row
file['claim bin'] = file['total claim amount'].apply(lambda row: c_bins(row))
file.head()
```

```python
sns.barplot(y=file["response_rate"], x=file["claim bin"],data=file)
```

```python
def clean(file):
    file = file.drop(['Unnamed: 0'], axis=1) #1
    file = file[file['State'].isna()==False] #2
    file = file[file['Months Since Last Claim'].isna()==False] #3
    file = file[file['Vehicle Class'].isna()==False] #4
    file['Vehicle Type'] = file['Vehicle Type'].fillna('B') #5
    file['Effective To Date'] = pd.to_datetime(file['Effective To Date']) #6
    file['Effective To Date Month'] = file['Effective To Date'].dt.month #7
    file[file['Effective To Date Month'].isin([1, 2, 3])] #8
    columns = [] #9
    for i in file.columns:
        columns.append(i.lower())
    file.columns = columns
    return file

```

##### Creating bins for total claim amount and plotting it

```python
# Describe the total claim amount so we can get the different percentile values
file['total claim amount'].describe()  
```

```python
# Use the describe values to create a function which will help put the row values in total claim amount to different bins
def c_bins(row):
    if row < 272:
        return ' < 25%' #Bin 1
    elif row < 383:
        return '25%-50%' #Bin 2
    elif row < 548:
        return '50%-75%' #Bin 3
    else:
        return '> 75%' #Bin 4
# Go and get every row of total claim amount and apply the bins() function to every row
file['claim bin'] = file['total claim amount'].apply(lambda row: c_bins(row))
file.head()
```

```python
#Sns barplot of response rate vs claim bins
sns.barplot(y=file["response_rate"], x=file["claim bin"],data=file)
```

##### Creating bins for income and plotting it

```python
file['income'].describe().T #To get descriptive stats
```

```python
#same function but for income

def i_bins(row):
    if row < 33890:
        return ' < 50%' #bin 1
    elif row < 62320:
        return '50%-75%' #bin 2
    else:
        return '>75%' #bin 3

# Go and get every row of income and apply the bins() function to every row
file['income bins'] = file['income'].apply(lambda row: i_bins(row))
file.head()
```

```python
#bar plot of response rate vs income bins

sns.barplot(y=file["response_rate"], x=file["income bins"],data=file)
```

```python
categoricals 
#Need to use Onehotencoder to convert categoricals into a new column 
#And assign a 1 or 0 (notation for true/false) value to the column so that it can be used
```

```python
#OneHotEncoder - Encodes categorical features as a one-hot numeric array

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(categoricals)  #encoder - drop first column of array for hot encoded categories because the first col can be infered using the others
encoded = encoder.transform(categoricals).toarray() #encoded - use encoder transform method to put categoricals into array
encoded_columns = OneHotEncoder(drop='first').fit(categoricals).get_feature_names(input_features=categoricals.columns) # this allows us to get the columns names for our encoded array

onehot_encoded_X = pd.DataFrame(encoded, columns = encoded_columns) #new onehotencoded columns X because its still features
onehot_encoded_X.head(20)
```

```python
new_X = pd.concat([onehot_encoded_X, min_max_X], axis=1) #concat transformed dataframes called new_X
#replace one_hot_encoded_x with new_X
```

##### Linear regression base model

```python
#This is the base model

lm = linear_model.LinearRegression() #linear regression model, lm is linear model
lm.fit(new_X,y) #we fit features and target to the model
print(lm.intercept_) #Intercept 
print(lm.coef_) #Coefficent for each column
r2_score(lm.predict(new_X),y)  #score is R2 score of the base model
```

## Train test split of data

The previous model has nothing to be tested on, it just gives us the values of intercept and coefficient so we cannot tell if this is good or not and thats why we will split the data

```python
# Dataset is split into 2 parts randomly so that I can create my model on training set and then check how well it does on the test set
# I will use test size as 0.20 because the dataset is less than 10000 rows, it makes no sense to use less data for training.

from sklearn.model_selection import train_test_split
# if you dont have random state in you train test split then your model is unreliable
X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size=0.20, random_state=42) # you can use whichever random state you want, the random state number allows two people to get the same sample so that we can test our analysis
```
